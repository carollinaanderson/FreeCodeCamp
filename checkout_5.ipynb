Esse dataset têm uma desproporcionalidade entre o quantitativo de homens e mulheres, sendo 15.547 homens e 12.354 mulheres.

Nesse dataset específico, que traz informações da Índia, então é interessante analisar o acesso das mulheres à educação para entender melhor se essa prevalência de homens com sintoma depressivo é somente uma questão relacionada à proporcionalidade.


* Relação entre Gênero e Ideação Suicida - apesar do gráfico fazer parecer visualmente que mais homens têm ideação suicida do que mulheres, na realidade ao avaliar os percentuais de resposta para cada variável é possível identificar que porporcionalemente. Para evitar esse problema do gráfico o que eu faço?

  - há mais mulheres estudantes com ideação suicida, sendo 63.347904% sendo que há 63.227631% de homens com ideação suicida





```python


# calcule o percentual de homens e mulheres para Have you ever had suicidal thoughts?

suicidal_thoughts_counts = df.groupby(['Gender', 'Have you ever had suicidal thoughts ?'])['id'].count().unstack(fill_value=0)
suicidal_thoughts_percentages = suicidal_thoughts_counts.div(suicidal_thoughts_counts.sum(axis=1), axis=0) * 100

# crie uma tabela com o resultado

result_table = pd.DataFrame({
    'Gender': suicidal_thoughts_percentages.index,
    'Suicidal Thoughts - Yes (%)': suicidal_thoughts_percentages['Yes'],
    'Suicidal Thoughts - No (%)': suicidal_thoughts_percentages['No'],
})
result_table
```
